```python
# 스크롤 내리면서 이미지 정보 다 담아주기
body = driver.find_element_by_tag_name('body')
# range() 가로 안에 스크롤 몇번 내릴 건지 지정! 
# 저같은 경우 불고기와 관련된 사진이 많이 없어서 100으로 설정 하여 최대한 많은 사진 담기로 결정
# 나중에 이미지 폴더에서 쓸데없는 사진들은 수기로 제거해줄 예정
soup = bs(driver.page_source, 'lxml')
for i in range(50):
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.3)
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.3)
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.3)
    # 아까 img_bulgogi 리스트 선언한 것을 여기에 계속 더해줌으로써 스크롤 내리면서 
    # 이미지 정보 계속 저장
    img_list += soup.select('img')

# 이미지의 직접적인 정보인 src 를 담아줄 리스트 선언
img_src = []

# 중복된 src 정보를 없애기 위해 new_img_src 리스트 선언
new_img_src= []

# 먼저 모든 src 정보를 img_src 리스트에 담는다.
for i in img_list :
    img_src.append(i['src'])
    
# img_src 리스트에 있는 중복된 정보를 제거해 주기 위해 아래와 같은 알고리즘 사용
for i in img_src:
    if i not in new_img_src:
        new_img_src.append(i)
        
# 마지막으로 중복 제거된 이미지 정보를 해당 폴더에 저장
for i in tq(range(len(new_img_src))):
    # ure(new_img_src[i], 'C:/Users/SMT068/Desktop/폴더명/'+'파일이름.'+str(i)+'.png')
    ure(new_img_src[i],"C:/Users/SMT021/foodimage/datasets/{}/{}.{}.png".format(foodname, foodname, cnt))
    cnt+=1
    time.sleep(0.2)
```
