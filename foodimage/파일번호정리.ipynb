```python
count = range(13000)
```

```python
import os, os.path, shutil, zipfile

fnames = []
exist_list = []

original_dataset_dir = "./datasets/"
food_dataset_dir = os.path.join(original_dataset_dir, "seamustardsoup")

fnames = ["seamustardsoup.{}.png".format(i) for i in count]

if not os.path.exists(food_dataset_dir+"new"):
    os.mkdir(food_dataset_dir+"new")
    
for fname in fnames:
    if os.path.exists(food_dataset_dir+"/"+fname):
        exist_list.append(fname)
        
for fname in exist_list:
    # 원래 이미지가 있는 폴더와 파읾명
    src = os.path.join(food_dataset_dir, fname)
    # 복사할 위치의 폴더와 파일명
    dst = os.path.join(food_dataset_dir+"new", "seamustardsoup.{}.png".format(exist_list.index(fname)))
    shutil.copyfile(src,dst)
```
